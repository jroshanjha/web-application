## Divided The dataset Into X and Y:- 

```python
# Prepare the data 
X = iris.drop(['species'], axis=1)  # 'sepal_width'
y = iris['species']
```

```python
models=[]
#models.append(('LinearRegression:',LinearRegression()))
models.append(('LR:',LogisticRegression(solver='liblinear',multi_class='ovr')))
models.append(('LDA:',LinearDiscriminantAnalysis()))
models.append(('KNN:',KNeighborsClassifier()))
models.append(('DTC(CART:)',DecisionTreeClassifier()))
models.append(('RFC: ',RandomForestClassifier(n_estimators=100)))
models.append(('NB:',GaussianNB()))
models.append(('Perceptron',Perceptron()))
models.append(('SVC',SVC())) # gamma='auto'
```

```python
seed = 41 # 6
scoring = 'accuracy'
```

## Split Data into training & testing Dataset:-

```python
# Split the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=seed) # stratify=None
```

```python
X_train.size,X_test.size
```

```python
y_train.size,y_test.size
```

```python
results =[]
names = []
for name,model in models:
    kfold = KFold(n_splits = 5, random_state = seed,shuffle=True)
    # n_splits = min(10, len(X_train))
    # kfold = KFold(n_splits=n_splits, random_state=seed, shuffle=True)
    cv_results = cross_val_score(model,X_train,y_train,cv= kfold,scoring =scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s : %f (%f)" %(name,cv_results.mean(),cv_results.std())
    print(msg)
```

## Comparing Tha Algorithms Scores:-

```python
fig = plt.figure()
fig.suptitle('Algoriths Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results,showmeans=True)
ax.set_xticklabels(names)
plt.show()
```

# LDA - Linear Discriminant Analaysis 

```python

model = LinearDiscriminantAnalysis()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

# Comprehensive Model Evaluation

```python

def evaluate_predictions(y_true, y_pred):
    results = {
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision (Macro)': precision_score(y_true, y_pred, average='micro'),
        'Recall (Macro)': recall_score(y_true, y_pred, average='micro'),
        'F1 Score (Macro)': f1_score(y_true, y_pred, average='micro')
    }
    return results

# ValueError: Target is multiclass but average='binary'. Please choose another average setting, one of [None, 'micro', 'macro', 'weighted'].
```

```python
evaluate_predictions(y_test,y_pred)
```

# Detailed Classification Report

```python

print("\nDetailed Classification Report:")
print(classification_report(y_test, y_pred))
```

```python

conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)
```
